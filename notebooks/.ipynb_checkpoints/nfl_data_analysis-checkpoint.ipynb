### Coming up with a Hypothesis
After looking at information regarding the defenses as a whole, I decided that the best course of action would be to target a specific group of defensive players. In my recent reading I have heard a common theme about NFL offenses getting faster and specifically that the wide receiver group is getting faster. I want to take a look at speed metrics surrounding the defensive position that most commonly guards the wide receiver group. With that in mind I will be looking at the Cornerback position.

The main question that I would like to answer is whether or not speed has a statistically significant effect on the outcome of a play. So I will look to split up all the Cornerbacks by average speed per play and average acceleration per play. I will then pit that against each players average EPA per play to determine if there is some effect. 

##### $h_0$ : There is no difference in EPA between the bottom 25% slowest CB's and the top 25% fastest CB's 
##### $h_a$: The top 25% fastest CB's will have a lower average EPA. 



### An explanation of EPA

Expected points added on the play, relative to the offensive team. Expected points is a metric that estimates the average of every next scoring outcome given the play's down, distance, yardline, and time remaining (numeric).  A lower EPA represents a poor performance on offense and a positive performance by the defense.
