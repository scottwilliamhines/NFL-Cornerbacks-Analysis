```python
week_list= [
week1,
week2,
week3,
week4,
week5,
week6,
week7,
week8,
week9,
week10,
week11,
week12,
week13,
week14,
week15,
week16,
week17,
]


def create_cb_df(lst):
    week_empty = pd.DataFrame({'nflId': [],
                           'displayName': [],
                           's': [],
                           'a': []
                          })
    for week in lst:
        week = week[week['position'] == "CB"][['s','a','nflId', 'displayName', 'playId']]\
        .groupby(['playId', 'nflId', 'displayName']).mean()\
        .reset_index(level = ['displayName', 'nflId', 'playId']).drop('playId', axis = 1)
        week_empty = pd.concat([week_empty,week], join= 'inner')
    return week_empty

cb_df = create_cb_df(week_list)
        
cb_df
```

```python
play_counts = pd.DataFrame(cb_df.groupby('nflId')['s'].count())
```

```python
play_counts.reset_index(level = 'nflId', inplace = True)
```

```python
play_counts.rename(columns = {'s': 'playCount'}, inplace = True)
play_counts
```

```python
epa = plays[plays['playType'] == 'play_type_pass'][['gameId','playId','epa']]
epa
```

```python
epa.set_index(['gameId', 'playId'], inplace=True)
```

```python
week_list= [
week1,
week2,
week3,
week4,
week5,
week6,
week7,
week8,
week9,
week10,
week11,
week12,
week13,
week14,
week15,
week16,
week17,
]


def create_cb_df(lst):
    week_empty = pd.DataFrame({
                            'gameId': [],
                            'playId': [],
                            'nflId': [],
                            'displayName': [],
                            's': [],
                            'a': []
                          })
    for week in lst:
        week = week[week['position'] == "CB"][['s','a','nflId', 'displayName', 'playId','gameId']]\
                    .groupby(['gameId','playId', 'nflId', 'displayName']).mean()\
                    .reset_index(level = ['gameId', 'playId', 'nflId', 'displayName'])
        week_empty = pd.concat([week_empty,week], join= 'inner')
    return week_empty

cb_avg_epa = create_cb_df(week_list)

cb_avg_epa.head()
```

```python
cb_avg_epa.set_index(['gameId', 'playId'], inplace=True)
```

```python
cb_avg_epa = cb_avg_epa.join(epa, on=['gameId','playId'])
cb_avg_epa.head()
```

```python
cb_df_final = cb_avg_epa.groupby(['nflId', 'displayName']).mean()\
            .reset_index(level = ['nflId','displayName'])
```

```python
cb_df_final.set_index('nflId', inplace = True)
```
